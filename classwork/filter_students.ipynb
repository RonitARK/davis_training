A file with name student.txt contains records in the format name, marks, city.
extract andd save only the student data whose marks is greater then 75 into a new file

```python
def filter_students(input_file, output_file):
    try:
        # Check read mode/existence by attempting to open
        with open(input_file, 'r') as infile:
            # Open destination file for writing
            with open(output_file, 'w') as outfile:

                for line in infile:
                    # Strip whitespace and split by comma
                    parts = line.strip().split(',')

                    # Ensure the line has all 3 parts (name, marks, city)
                    if len(parts) == 3:
                        name, marks_str, city = parts

                        # Convert marks to a number for comparison
                          marks = float(marks_str)
                          if marks > 75:
                              # Write the valid record to the new file
                              outfile.write(f"{name}, {marks}, {city}\n")

        print(f"Extraction complete. Check '{output_file}' for results.")

    except FileNotFoundError:
        print(f"Error: '{input_file}' not found.")
    except PermissionError:
        print(f"Error: Permission denied for file access.")

# Run the function
filter_students('student.txt', 'high_achievers.txt')
```
