A Program to copy the content of one file to another

```python
def copy_file_manual(source, destination):
    try:
        # Checking read mode by attempting to open the source
        with open(source, 'r') as src_file:
            # If we reach here, the file is readable
            content = src_file.read()

            # Writing the content to the destination
            with open(destination, 'w') as dest_file:
                dest_file.write(content)

        print(f"Successfully copied '{source}' to '{destination}'.")

    except FileNotFoundError:
        print(f"Error: The file '{source}' was not found.")
    except PermissionError:
        print(f"Error: You do not have permission to read '{source}'.")

# Usage
copy_file_manual('origin.txt', 'backup.txt')
```
