```python
def find_greatest_digit(number):

    num_str = str(number)
    max_digit = -1

    for digit_char in num_str:
        digit = int(digit_char)

        if digit > max_digit:
            max_digit = digit

    return max_digit


number = int(input("Enter the Number: "))
result = find_greatest_digit(number)
print(f"The greatest digit in {number} is: {result}")
```
