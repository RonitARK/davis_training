WAP python to create a list of any 10 numbers and sort them in descending order without using any library method for sorting

```python

numbers_bubble = [10,25,54,84,56,1,2,0,14,55]
print("Original list (Bubble Sort):", numbers_bubble)

# Sort the list in descending order using Bubble Sort
n = len(numbers_bubble)
for i in range(n - 1):
    for j in range(0, n - i - 1):
        if numbers_bubble[j] < numbers_bubble[j + 1]:
            numbers_bubble[j], numbers_bubble[j + 1] = numbers_bubble[j + 1], numbers_bubble[j]

print("Sorted list (descending - Bubble Sort):", numbers_bubble)
```
