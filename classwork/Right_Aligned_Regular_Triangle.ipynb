```python
def right_aligned_standard(rows):
    """Prints a right-aligned, right-side-up triangle."""
    for i in range(1, rows + 1):
        spaces = " " * (rows - i)
        stars = "*" * i
        print(spaces + stars)

# Call the function
right_aligned_standard(5)
```
