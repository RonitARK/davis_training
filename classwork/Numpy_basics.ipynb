```python
import numpy as np

def get_even_factors_simple(num):
    even_factors = []
    if num == 0:
        return even_factors # Zero has no even factors in this context

    # Get factors for the absolute value of the number
    abs_num = abs(num)
    for i in range(1, abs_num + 1):
        if abs_num % i == 0 and i % 2 == 0:
            even_factors.append(i)
    return even_factors

# Test with different types of numbers
numbers_to_test = [24, -8, 0, 15, 100, -50]

for number_to_check in numbers_to_test:
    even_factors_list = get_even_factors_simple(number_to_check)

    if even_factors_list:
        even_factors_array = np.array(even_factors_list)
        print(f"The even factors of {number_to_check} are: {even_factors_array}")
    elif number_to_check == 0:
        print(f"Zero has no even factors.")
    else:
        print(f"There are no even factors for {number_to_check} (or all factors are odd).")
```
