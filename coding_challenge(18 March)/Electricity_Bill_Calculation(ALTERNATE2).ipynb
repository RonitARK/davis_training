Question 13: Calculate electricity bill based on units consumed.

Input: Enter units

Output: Total bill amount

```python
units = float(input("Enter units: "))
rate_per_unit = 5.0 # Assuming a flat rate of $5 per unit
total_bill = units * rate_per_unit
print(f"Total bill amount: ${total_bill}")
```
