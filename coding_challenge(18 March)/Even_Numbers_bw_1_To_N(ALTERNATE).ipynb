Question 9: Print all even numbers between 1 to N.

Input: Enter N

Output: Even numbers list

```python
def get_evens(number):
    evens = []
    for i in range(1, number + 1):
        if i % 2 == 0:
            evens.append(i)
    return evens

number = int(input("Enter Number: "))
print("Even numbers list:", get_evens(number))
```
