Question 8: Calculate electricity bill based on units consumed.

Input: Enter units

Output: Total bill amount

```python
def calculate_electricity_bill(units):
    bill = 0.0
    if units <= 50:
        bill = units * 3.50
    elif units <= 150:
        # First 50 units
        bill += 50 * 3.50
        # Next units (from 51 to 150)
        bill += (units - 50) * 4.00
    elif units <= 250:
        # First 50 units
        bill += 50 * 3.50
        # Next 100 units (from 51 to 150)
        bill += 100 * 4.00
        # Next units (from 151 to 250)
        bill += (units - 150) * 5.20
    else:
        # First 50 units
        bill += 50 * 3.50
        # Next 100 units (from 51 to 150)
        bill += 100 * 4.00
        # Next 100 units (from 151 to 250)
        bill += 100 * 5.20
        # Remaining units (above 250)
        bill += (units - 250) * 6.50
    return bill

# Get units consumed from the user

units_consumed = float(input("Enter units consumed: "))
if units_consumed < 0:
  print("Units consumed cannot be negative.")
else:
  total_bill = calculate_electricity_bill(units_consumed)
print(f"Total electricity bill for {units_consumed} units: Rs. {total_bill:.2f}")
```
