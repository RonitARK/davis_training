Question 3: Calculate electricity bill based on units consumed.

Input: Enter units

Output: Total bill amount

```python
def calculate_bill(units):
    if units <= 50:
        bill = units * 0.50
    elif units <= 150:
        bill = (50 * 0.50) + ((units - 50) * 0.75)
    elif units <= 250:
        bill = (50 * 0.50) + (100 * 0.75) + ((units - 150) * 1.20)
    else:
        bill = (50 * 0.50) + (100 * 0.75) + (100 * 1.20) + ((units - 250) * 1.50)

    # Add a fixed charge for example, or any other additional charges
    fixed_charge = 50.00
    total_bill = bill + fixed_charge
    return total_bill

# Get units consumed from the user
while True:
    try:
        units_consumed = float(input("Enter units consumed: "))
        if units_consumed < 0:
            print("Units cannot be negative. Please enter a positive value.")
        else:
            break
    except ValueError:
        print("Invalid input. Please enter a numeric value for units.")

# Calculate and display the total bill
bill_amount = calculate_bill(units_consumed)
print(f"Total electricity bill for {units_consumed} units: ${bill_amount:.2f}")
```
