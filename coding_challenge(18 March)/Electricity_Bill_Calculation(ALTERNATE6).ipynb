```python
def calculate_electricity(units):
    bill = 0
    if units <= 100:
        bill = units * 1.5  # First 100 units at $1.5
    elif units <= 200:
        bill = (100 * 1.5) + ((units - 100) * 2.5) # Next 100 at $2.5
    else:
        bill = (100 * 1.5) + (100 * 2.5) + ((units - 200) * 4.0) # Rest at $4.0
    return bill

units = float(input("Enter units: "))
print(f"Total bill amount: ${calculate_electricity(units)}")
```
