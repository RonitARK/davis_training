```python
number = int(input("Enter Number: "))
if number < 1:
  print("Please enter a positive integer for N.")
else:
  even_numbers = []
  for i in range(1, number + 1):
    if i % 2 == 0:
      even_numbers.append(i)
print("Even numbers list:", even_numbers)
```
