```python
def find_even_factors(number):
    even_factors = []
    for i in range(1, number + 1):
        if number % i == 0 and i % 2 == 0:
            even_factors.append(i)

    if even_factors:
        print(f"The even factors of {number} are: {even_factors}")
    else:
        print(f"There are no even factors for {number}.")

number = int(input("Enter the Number:"))
result = find_even_factors(number)
```
